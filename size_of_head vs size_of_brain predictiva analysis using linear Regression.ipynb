```python
#importing the required packages
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)
```

```python
#importing the dataset
data=pd.read_csv('headbrain.csv')
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
#mean of X and Y
mean_x=np.mean(X)
mean_y=np.mean(Y)
n=len(X)
num=0
den=0
for i in range(n):
    num+=(X[i]-mean_x)*(Y[i]-mean_y)
    den+=(X[i]-mean_x)**2
    b1=num/den
    b0=mean_y-(b1*mean_x)

    
```

```python
print(b1,b0)
```

```python
#plt
max_x=np.max(X)+100
min_x=np.min(X)-100
x=np.linspace(max_x,min_x,1000)
y=b0+b1*x
plt.plot(x,y,color="orange",label="LinearRegression analysis")
plt.scatter(x,y,color="blue",label="scatterplot")
plt.xlabel("head size")
plt.ylabel("brain size")
plt.plot()
```

```python
#R square method ===> summation(y_prediction-y_mean)**2/summation(y-y_mean)**2
ss_t=0
ss_r=0
for i in range(n):
    y_pred=(b0+b1)*X[i]
    ss_t+=(Y[i]-mean_y)**2
    ss_r+=(Y[i]-y_pred)**2
    r2=1-ss_r/ss_t
```

```python
print(r2)
```

```python
#creating the linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X=X.reshape((n,1))
reg=LinearRegression()
reg=reg.fit(X,Y)
Y_pred=reg.predict(X)
r2_score=reg.score(X,Y)
```
